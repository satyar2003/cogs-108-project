### Analysis of books on a celebrity booklist

The first graph shows the average ranking of books on a celebrity booklist compared to books that were not on a celebrity booklist.

The second graph shows the average number of weeks on the NYT bestseller list books on a celebrity booklist were compared to the average number of weeks books that were not on a celebrity booklist were.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Bookinfo.csv')
titles_on_booklist = df.loc[df['On a Celebrity Reading List'] == True]['Title'].unique().tolist()
titles_not_on_booklist = df.loc[df['On a Celebrity Reading List'] == False]['Title'].unique().tolist()

avg_score_booklist = []
for title in titles_on_booklist:
    x = df.loc[df['Title'] == title]['Rank'].tolist()
    avg_score_booklist.append(sum(x) / len(x))
avg_score_no_booklist = []
for title in titles_not_on_booklist:
    x = df.loc[df['Title'] == title]['Rank'].tolist()
    avg_score_no_booklist.append(sum(x) / len(x))

df2 = pd.DataFrame(data={'On Booklist': ['Yes', 'No'], 
                         'Average Ranking': [sum(avg_score_booklist) / len(avg_score_booklist), sum(avg_score_no_booklist) / len(avg_score_no_booklist)]})
sns.set_theme(rc={'figure.figsize':(11.7,8.27)}, font_scale=1)
sns.barplot(data=df2, x='On Booklist', y='Average Ranking')
plt.title('Average Ranking of Books on Celebrity Booklist vs. Books not on Celebrity Booklist')
```

The barplot shows that there is not much difference in average ranking based on a book's apperance on a celebrity booklist. It seems there's about a difference of 1 ranking spot between the two categories, where being on a celebrity booklist results in a higher rank on average. However, it should be noted that there were only 8 unique books found on a celebrity booklist out of the total 213 unique books on the NYT bestseller list.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Bookinfo.csv')
titles_on_booklist = df.loc[df['On a Celebrity Reading List'] == True]['Title'].unique().tolist()
titles_not_on_booklist = df.loc[df['On a Celebrity Reading List'] == False]['Title'].unique().tolist()

weeks_on_nyt_booklist = []
for title in titles_on_booklist:
    x = df.loc[df['Title'] == title]['Max Weeks on List'].tolist()
    weeks_on_nyt_booklist.append(sum(x) / len(x))
weeks_on_nyt_no_booklist = []
for title in titles_not_on_booklist:
    x = df.loc[df['Title'] == title]['Max Weeks on List'].tolist()
    weeks_on_nyt_no_booklist.append(sum(x) / len(x))

df3 = pd.DataFrame(data={'On Booklist': ['Yes', 'No'],
                         'weeks': [sum(weeks_on_nyt_booklist) / len(weeks_on_nyt_booklist), sum(weeks_on_nyt_no_booklist) / len(weeks_on_nyt_no_booklist)]})
sns.barplot(data=df3, x='On Booklist', y='weeks')
plt.title('Average Number of Weeks on NYT Bestseller List for Books on Celebrity Booklist vs. Book not on Celebrity Booklist')
plt.ylabel('Average Number of Weeks on NYT Bestseller List')
```

The barplot shows that there is a significant difference regarding the average number of weeks books have been on the NYT Bestseller list when differentiating them based on their presence on a celebrity booklist. It seems that most of the NYT Bestsellers that were found on a celebrity booklist were only recent additions to the Bestseller list, with some possibly being removed from the list soon after their debut. As mentioned with the previous barplot, the number of unique books found on a celebrity booklist is a small fraction of the total number of unique books on the Bestseller list, which may cause some large distinctions as shown above.
